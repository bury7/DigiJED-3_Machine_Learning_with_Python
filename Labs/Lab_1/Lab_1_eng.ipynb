#### **Task 8. Print the same statistics for the qualitative features, but within each class. Are there any qualitative attributes that have one unique value within a particular class?**

The data was grouped by class using the `pivot_table()` method, and the unique values in the classes were displayed, as well as the frequency of each unique value in the classes.

```python
columns = ["workclass", "education", "marital_status", "occupation", "relationship", "race", "sex", "native_country"]
for column in columns:
    display(df.pivot_table(index="gross_income", values=column, aggfunc=lambda x: x.unique()))
    display(df.pivot_table(index=column, columns="gross_income", aggfunc='size', fill_value=0))
```

#### **Task 10. If there are any gaps in the data set, remove or fill them in.**

Using the `dropna()` method, the rows with gaps in the `workclass` and `native_country` columns are deleted.

```python
df.dropna(subset=["workclass", "native_country"], inplace=True)
```

#### **Task 11: Perform the transformation of qualitative features. You need to master two types of transformations: converting each qualitative feature into a set of binary features (damme variables) and recoding the values of the feature into numbers. What is the purpose of these transformations?**

Converted the qualitative nominal feature `sex` into a set of binary features using the `get_dummies()` function. The `drop_first` parameter is set to "True" in order to delete the first created category, but the information about the first category is not lost, since its value can be easily restored from the second column.

```python
df = pd.get_dummies(df, columns=['sex'], drop_first=True)
```

The values of the qualitative sequential feature `education` are recoded into numbers using the `map()` method.

```python
df["education"] = df["education"].map({"Preschool": 1, "1st-4th": 2, "5th-6th": 3, "7th-8th": 4, "9th": 5, "10th": 6, "11th": 7, "12th": 8, "Some-college": 9, 
                  "HS-grad": 10, "Assoc-acdm": 11, "Assoc-voc": 12, "Bachelors": 13, "Masters": 14, "Prof-school": 15, "Doctorate": 16})
```

#### **Task 12. Demonstrate the possibility of normalization and standardization of quantitative features. What is the purpose of these transformations? Is it always necessary to perform them?**

The `fnlwgt` feature has been normalized.

```python
(df["fnlwgt"] - df["fnlwgt"].min()) / (df["fnlwgt"].max() - df["fnlwgt"].min())
```

The `fnlwgt` feature has been standardized.

```python
(df["fnlwgt"] - df["fnlwgt"].mean()) / df["fnlwgt"].std(ddof=1)
```

Normalization and standardization are performed to bring data to the same scale and facilitate analysis and modeling. Their implementation is not always mandatory, but is often recommended to improve the efficiency and stability of algorithms.

#### **Task 13. Remove from the dataset features that are definitely not useful in the future (e.g., id, features with one unique value or containing more than 50% missing values). Explain why you are removing the selected features.**

The `education` and `native_country` attributes were removed from the dataset using the `drop()` method. The `education` feature was dropped because information about educational level is already presented in the `education_num` column, which avoids redundancy of information and can improve model performance by avoiding duplication of data. As for the `native_country` feature, it was removed because most individuals have a value of `United-States` and therefore this feature does not add much information to distinguish individuals in classes, helping to avoid excessive redundancy or insufficient variation that could lead to overtraining the model on a particular subset of the data.

```python
df.drop(columns=["education", "native_country"], inplace=True)
```
